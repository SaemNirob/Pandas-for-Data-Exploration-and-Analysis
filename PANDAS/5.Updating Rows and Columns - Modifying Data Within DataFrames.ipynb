### Now let's talk about apply(), applymap(), map(), and replace()

##### 1.apply(): Works on whole rows or columns (DataFrame) or elements in a Series.
##### 2.applymap(): Works on each individual value in a DataFrame.
##### 3.map(): Works on each element in a Series. It maps values based on a dictionary or function. If a value does not match, it returns NaN.
##### 4.replace(): Replaces specific values or handles missing values (NaN) in a DataFrame or Series.
Note:

##### - Use apply() when you need to work on rows or columns.

##### - Use applymap() when you need to work on every single value in a DataFrame.

##### - Use map() when you need to map values in a Series. If a value does not match, it returns NaN.

##### - Use replace() to replace specific values or handle missing data (NaN) in DataFrames or Series.
