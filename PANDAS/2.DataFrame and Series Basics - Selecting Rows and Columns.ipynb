#### The main difference between loc and iloc is:

##### loc: 
 - Accesses data by labels (row and column names).

##### iloc:
 - Accesses data by index position (integer-based, like row/column numbers).

```python
# Create a dictionary with mixed data types
data = {
    'Name': ['Rashedul', 'Zannat', 'Rishan'],
    'Age': [27, 24, 22],
    'Occupation': ['Data Scientist', 'Software Engineer', 'Teacher'],
    'Salary': [80000, 50000, 70000]
}

# Create DataFrame
df2 = pd.DataFrame(data)

# Display the first few rows of the DataFrame
df2.head()
```

```python
df2.set_index('Name', inplace = True)
df2
```

```python
df2.loc[0]
# It will give an error because there is no index value called '0'.
```

```python
df2.loc['Zannat']
```

```python
df2.iloc[0]
# It will work because iloc works with index position, not value.
```
