```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.replace('NA', np.nan, inplace = True)
df.replace('Missing', np.nan,inplace = True)
df
```

```python
df.dropna()
# by default: df.dropna(axis = 0, how = 'any')  
```

```python
df.dropna(axis = 0, how = 'any')
```

```python
df.dropna(axis = 0, how = 'all')
```

```python
df.dropna(axis = 1, how = 'all')
```

```python
df.dropna(axis = 1, how = 'any')
```

##### Whatever we use in the 'how', 'any'/'all' will both give the same result because of the subset.

```python
df.dropna(axis = 0, how = 'any', subset = ['email'])
```

##### In the below example: df.dropna(axis=0, how='all', subset=['first', 'email']), the how='all' parameter is used as 'AND' logic.

```python
df.dropna(axis = 0, how = 'all', subset = ['first', 'email'])
```

##### In the below example: df.dropna(axis=0, how='any', subset=['first', 'email']), the how='all' parameter is used as 'OR' logic.

```python
df.dropna(axis = 0, how = 'any', subset = ['first', 'email'])
```
